Detectar y tratar valores atipicos

IQR

```python
import pandas as pd

def detect_outliers_iqr(df, column):
    """
    Detecta outliers en una columna usando el método IQR.
    
    Retorna:
        - outliers: DataFrame con los valores fuera del rango permitido
        - low_limit, high_limit: límites usados para filtrar
    """
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1

    low_limit = Q1 - 1.5 * IQR
    high_limit = Q3 + 1.5 * IQR
    
    outliers = df[(df[column] < low_limit) | (df[column] > high_limit)]
    
    return outliers, low_limit, high_limit
```

```python
outliers_LB, low, high = detect_outliers_iqr(df, "LB")
outliers_LB.head()
```

```python
def treat_outliers_iqr(df, column, method="cap"):
    """
    Trata outliers de una columna usando IQR.
    
    method:
        - "cap": sustituye outliers por los límites (winsorization)
        - "remove": elimina las filas con outliers
    """
    outliers, low_limit, high_limit = detect_outliers_iqr(df, column)

    if method == "cap":
        df[column] = df[column].clip(lower=low_limit, upper=high_limit)

    elif method == "remove":
        df = df[(df[column] >= low_limit) & (df[column] <= high_limit)]

    return df
```

```python
df = treat_outliers_iqr(df, "LB", method="cap")
```

Z-Score

```python
import numpy as np

def detect_outliers_zscore(df, column, threshold=3):
    """
    Detecta outliers usando Z-Score.
    threshold = 3 es lo estándar.
    """
    mean = df[column].mean()
    std = df[column].std()

    z_scores = (df[column] - mean) / std

    outliers = df[np.abs(z_scores) > threshold]

    return outliers
```

```python
detect_outliers_zscore(df, "LB").head()
```

```python
def treat_outliers_zscore(df, column, threshold=3, method="remove"):
    """
    Trata outliers basados en Z-score.
    
    method:
        - "remove": elimina los outliers
        - "cap": aplica winsorization usando los límites threshold*std
    """
    mean = df[column].mean()
    std = df[column].std()
    
    upper_limit = mean + threshold * std
    lower_limit = mean - threshold * std

    if method == "remove":
        df = df[(df[column] >= lower_limit) & (df[column] <= upper_limit)]
    
    elif method == "cap":
        df[column] = df[column].clip(lower=lower_limit, upper=upper_limit)
    
    return df
```

```python
df = treat_outliers_zscore(df, "AC", threshold=3, method="remove")
```

2. Análisis de Datos

```python
import pandas as pd

def check_data_completeness_numer(df: pd.DataFrame) -> pd.DataFrame:
    """
    Analiza columnas numéricas para revisar:
    - Conteo de nulos
    - Porcentaje de completitud
    - Tipo de dato
    - Estadísticos de dispersión
    - Clasificación en continuas o discretas

    Return:
        DataFrame resumen.
    """

    numer_cols = df.select_dtypes(include=['int64', 'float64']).columns

    resultados = []

    for col in numer_cols:
        series = df[col]

        n_nulos = series.isnull().sum()
        porcentaje_completitud = 1 - (n_nulos / len(series))

        dispersion = {
            "min": series.min(),
            "max": series.max(),
            "std": series.std(),
            "var": series.var(),
            "iqr": series.quantile(0.75) - series.quantile(0.25)
        }

        # Clasificación
        n_unique = series.nunique()
        if n_unique > 10:
            tipo = "Continua"
        else:
            tipo = "Discreta"

        resultados.append({
            "columna": col,
            "n_nulos": n_nulos,
            "porcentaje_completitud": porcentaje_completitud,
            "dtype": str(series.dtype),
            "min": dispersion["min"],
            "max": dispersion["max"],
            "std": dispersion["std"],
            "var": dispersion["var"],
            "iqr": dispersion["iqr"],
            "clasificacion": tipo
        })

    return pd.DataFrame(resultados)
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

def plot_histogram(
    df: pd.DataFrame,
    column: str,
    group: str = None,
    bins: int = 30,
    kde: bool = True,
    figsize: tuple = (8,5)
):
    """
    Histograma con opción de agrupar por categoría y añadir KDE.
    
    Args:
        df: DataFrame.
        column: Columna numérica a graficar.
        group: Columna categórica para dividir datos por color.
        bins: Número de bins del histograma.
        kde: Mostrar KDE (línea de densidad).
        figsize: Tamaño de la figura.
    """

    plt.figure(figsize=figsize)

    if group:
        sns.histplot(
            data=df,
            x=column,
            hue=group,
            bins=bins,
            kde=kde,
            element="step",
            stat="density",
            common_norm=False
        )
    else:
        sns.histplot(
            data=df,
            x=column,
            bins=bins,
            kde=kde,
            color="skyblue",
            stat="density"
        )

    plt.title(f"Histograma de {column} {'por ' + group if group else ''}")
    plt.xlabel(column)
    plt.ylabel("Densidad")
    plt.grid(True, alpha=0.3)
    plt.show()
```

```python
import pandas as pd
import plotly.express as px

def plot_histogram_interactive(
    df: pd.DataFrame,
    column: str,
    group: str = None,
    bins: int = 30
):
    """
    Histograma interactivo con Plotly para exploración dinámica.
    
    Args:
        df: DataFrame.
        column: Columna numérica a graficar.
        group: Columna categórica para colorear.
        bins: Número de bins.
    """

    fig = px.histogram(
        df,
        x=column,
        color=group,
        nbins=bins,
        marginal="rug",   # línea bajo el histograma
        opacity=0.75
    )

    fig.update_layout(
        title=f"Histograma interactivo de {column}",
        xaxis_title=column,
        yaxis_title="Frecuencia",
        bargap=0.05
    )

    fig.show()
```

* Boxplots

```python
import pandas as pd
import plotly.express as px
from plotly.subplots import make_subplots
import plotly.graph_objects as go

def boxplot_by_class(
    df: pd.DataFrame,
    numeric_col: str,
    class_col: str
):
    """
    Genera boxplots en subplots por clase objetivo usando Plotly.

    Args:
        df: DataFrame
        numeric_col: Columna numérica a analizar
        class_col: Columna categórica que define las clases (objetivo)
    """

    classes = df[class_col].unique()
    n_classes = len(classes)

    # Crear subplots
    fig = make_subplots(
        rows=1,
        cols=n_classes,
        subplot_titles=[f"{class_col}: {c}" for c in classes]
    )

    # Agregar cada boxplot a su subplot
    for i, cls in enumerate(classes, start=1):
        subset = df[df[class_col] == cls]
        fig.add_trace(
            go.Box(
                y=subset[numeric_col],
                name=str(cls),
                boxmean=True
            ),
            row=1,
            col=i
        )

    fig.update_layout(
        title=f"Boxplots de {numeric_col} por {class_col}",
        height=500,
        width=300 * n_classes,
        showlegend=False
    )

    fig.show()
```

```python
boxplot_by_class(df, numeric_col="LB", class_col="NSP")

```

* Barras Horizontales

```python
import seaborn as sns
import matplotlib.pyplot as plt

def plot_horizontal_bar_pro(df, column, top_n=None):
    freq = df[column].value_counts()

    if top_n:
        freq = freq.head(top_n)

    plt.figure(figsize=(10, 6))
    sns.set_theme(style="whitegrid")

    sns.barplot(
        y=freq.index,
        x=freq.values,
        orient="h",
        palette="crest"
    )

    plt.xlabel("Frecuencia", fontsize=12)
    plt.ylabel(column, fontsize=12)
    plt.title(f"Distribución de {column}", fontsize=14, weight="bold")
    plt.tight_layout()
    plt.show()
```

```python
def plot_horizontal_bar_labels(df, column, top_n=None):
    freq = df[column].value_counts()

    if top_n:
        freq = freq.head(top_n)

    plt.figure(figsize=(10, 6))
    sns.set_theme(style="ticks")

    ax = sns.barplot(
        y=freq.index,
        x=freq.values,
        orient="h",
        palette="viridis"
    )

    # Etiquetas al final de cada barra
    for i, v in enumerate(freq.values):
        ax.text(v + (v*0.01), i, str(v), va='center')

    plt.xlabel("Frecuencia")
    plt.ylabel(column)
    plt.title(f"Distribución de {column}", fontsize=14, weight="bold")
    plt.tight_layout()
    plt.show()
```

```python
def plot_horizontal_bar_percentage(df, column, top_n=None):
    freq = df[column].value_counts(normalize=True) * 100

    if top_n:
        freq = freq.head(top_n)

    plt.figure(figsize=(10, 6))
    sns.set_theme(style="whitegrid")

    ax = sns.barplot(
        y=freq.index,
        x=freq.values,
        orient="h",
        palette="mako"
    )

    for i, v in enumerate(freq.values):
        ax.text(v + 1, i, f"{v:.1f}%", va="center")

    plt.xlabel("Porcentaje (%)")
    plt.ylabel(column)
    plt.title(f"Distribución porcentual de {column}", fontsize=14, weight="bold")
    plt.tight_layout()
    plt.show()
```

```python
def plot_horizontal_bar_top(df, column, top_n=10):
    freq = df[column].value_counts().head(top_n).sort_values()

    plt.figure(figsize=(10, 6))
    sns.set_theme(style="whitegrid")

    ax = sns.barplot(
        y=freq.index,
        x=freq.values,
        orient="h",
        palette="rocket"
    )

    for i, v in enumerate(freq.values):
        ax.text(v + 0.5, i, str(v), va="center")

    plt.xlabel("Frecuencia")
    plt.ylabel(column)
    plt.title(f"Top {top_n} categorías de {column}", fontsize=14, weight="bold")
    plt.tight_layout()
    plt.show()
```

```python
plot_horizontal_bar_pro(df, "CLASS")
plot_horizontal_bar_labels(df, "NSP")
plot_horizontal_bar_percentage(df, "AC")
plot_horizontal_bar_top(df, "CLASS", top_n=5)
```

* Gráficos de Líneas

```python
import pandas as pd
import matplotlib.pyplot as plt

def plot_line(df: pd.DataFrame, numeric_column: str, time_column: str = None):
    """
    Grafica un gráfico de línea simulando serie temporal.
    
    Args:
        df: DataFrame con los datos.
        numeric_column: Columna numérica a graficar.
        time_column: Columna que representa el tiempo (opcional). Si no se da, se usa el índice.
    """
    
    plt.figure(figsize=(12, 5))
    
    if time_column and time_column in df.columns:
        x = df[time_column]
    else:
        x = df.index  # usar el índice como “tiempo”
    
    y = df[numeric_column]
    
    plt.plot(x, y, marker='o', linestyle='-', color='teal')
    plt.xlabel("Tiempo" if not time_column else time_column)
    plt.ylabel(numeric_column)
    plt.title(f"Serie temporal simulada de {numeric_column}")
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```

```python
# Usando índice como tiempo
plot_line(df, numeric_column="AC")

# Si tuvieras una columna tipo fecha o tiempo
# plot_line(df, numeric_column="AC", time_column="Date")
```

* Dot Plots

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

def plot_dot_plot(df: pd.DataFrame, numeric_column: str, group_column: str):
    """
    Genera un dot plot comparando dos o más grupos.
    
    Args:
        df: DataFrame con los datos.
        numeric_column: Columna numérica a graficar.
        group_column: Columna categórica para agrupar.
    """
    
    plt.figure(figsize=(10, 6))
    sns.stripplot(x=group_column, y=numeric_column, data=df, jitter=True, palette="Set2")
    
    plt.xlabel(group_column)
    plt.ylabel(numeric_column)
    plt.title(f"Dot Plot de {numeric_column} por {group_column}")
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.tight_layout()
    plt.show()
```

```python
# Ejemplo: comparar la columna 'AC' entre clases 'CLASS'
plot_dot_plot(df, numeric_column="AC", group_column="CLASS")
```

* Densidad

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

def plot_density_by_class(df: pd.DataFrame, numeric_column: str, class_column: str):
    """
    Genera un gráfico de densidad (KDE) para una variable numérica, comparando múltiples clases.
    
    Args:
        df: DataFrame con los datos.
        numeric_column: Columna numérica a graficar.
        class_column: Columna categórica para diferenciar las clases.
    """
    
    plt.figure(figsize=(10, 6))
    sns.kdeplot(data=df, x=numeric_column, hue=class_column, fill=True, alpha=0.5, palette="Set2")
    
    plt.xlabel(numeric_column)
    plt.ylabel("Densidad")
    plt.title(f"Gráfico de densidad de {numeric_column} por {class_column}")
    plt.grid(True, linestyle='--', alpha=0.5)
    plt.tight_layout()
    plt.show()
```

```python
# Comparar la distribución de 'AC' entre las clases 'CLASS'
plot_density_by_class(df, numeric_column="AC", class_column="CLASS")
```

* Violin

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

def plot_violin_swarm(df: pd.DataFrame, numeric_column: str, class_column: str):
    """
    Genera un violin plot de una variable numérica por clase,
    con overlay de swarmplot para ver los puntos individuales.
    
    Args:
        df: DataFrame con los datos.
        numeric_column: Columna numérica a graficar.
        class_column: Columna categórica para diferenciar las clases.
    """
    
    plt.figure(figsize=(10, 6))
    
    # Violin plot
    sns.violinplot(x=class_column, y=numeric_column, data=df, palette="Set2", inner=None, alpha=0.7)
    
    # Overlay con swarmplot
    sns.swarmplot(x=class_column, y=numeric_column, data=df, color="k", alpha=0.6)
    
    plt.xlabel(class_column)
    plt.ylabel(numeric_column)
    plt.title(f"Violin Plot con Swarm de {numeric_column} por {class_column}")
    plt.grid(axis='y', linestyle='--', alpha=0.5)
    plt.tight_layout()
    plt.show()
```

```python
# Ejemplo: visualizar 'AC' por 'CLASS'
plot_violin_swarm(df, numeric_column="AC", class_column="CLASS")
```

* Heatmap

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

def plot_correlation_heatmap(df: pd.DataFrame, method: str = 'pearson'):
    """
    Genera un heatmap de correlación para todas las variables numéricas del DataFrame.
    
    Args:
        df: DataFrame con los datos.
        method: Método de correlación ('pearson', 'spearman', 'kendall').
    """
    
    # Seleccionamos solo columnas numéricas
    numeric_df = df.select_dtypes(include='number')
    
    # Calculamos la matriz de correlación
    corr = numeric_df.corr(method=method)
    
    # Plot
    plt.figure(figsize=(12, 10))
    sns.heatmap(corr, annot=True, fmt=".2f", cmap="coolwarm", linewidths=0.5)
    
    plt.title(f"Heatmap de correlación ({method})")
    plt.tight_layout()
    plt.show()
```
